```python
import numpy as np
def softmax(x):
    e_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return e_x / e_x.sum(axis=-1, keepdims=True)
sentence = "What are the symptoms of diabetes?"
tokens   = sentence.lower().replace("?","").split()
seq_len  = len(tokens)   
d_model  = 8             
d_k      = 4             
np.random.seed(42)
X = np.random.randn(seq_len, d_model)
W_Q = np.random.randn(d_model, d_k)
W_K = np.random.randn(d_model, d_k)
W_V = np.random.randn(d_model, d_k)
Q = X @ W_Q
K = X @ W_K
V = X @ W_V
scores = Q @ K.T / np.sqrt(d_k)
attention_weights = softmax(scores)
output = attention_weights @ V
print("Tokens:", tokens)
print("\nRaw Attention Scores (Q·Kᵀ / √d_k):")
print(np.round(scores, 4))
print("\nAttention Weights (after softmax):")
print(np.round(attention_weights, 4))
print("\nSelf-Attention Output:")
print(np.round(output, 4))
print("\nPer-token attention distribution")
for i, tok in enumerate(tokens):
    top = np.argsort(attention_weights[i])[::-1][:3]
    print(f"{tok:12s} attends most to: " +
          ", ".join(f"{tokens[j]}({attention_weights[i,j]:.3f})" for j in top))
```
