```python
import math

def sigmoid(x):
    return 1/(1+math.exp(-x))

def tanh(x):
    return math.tanh(x)

# input sequence
x = [1,2,3,4]

# weights
Wf, Whf = 0.5, 0.1
Wi, Whi = 0.6, 0.2
Wc, Whc = 0.7, 0.3
Wo, Who = 0.8, 0.4

# initial states
h = 0
c = 0

for t in range(len(x)):

    xt = x[t]
    print("\n====================")
    print("Time step",t+1,"  x =",xt)

    # Forget Gate
    zf = Wf*xt + Whf*h
    f = sigmoid(zf)

    print("\nForget Gate:")
    print("zf = (0.5*",xt,") + (0.1*",round(h,4),") =",round(zf,4))
    print("f = sigmoid(",round(zf,4),") =",round(f,4))

    # Input Gate
    zi = Wi*xt + Whi*h
    i = sigmoid(zi)

    print("\nInput Gate:")
    print("zi = (0.6*",xt,") + (0.2*",round(h,4),") =",round(zi,4))
    print("i = sigmoid(",round(zi,4),") =",round(i,4))

    # Candidate Cell
    zc = Wc*xt + Whc*h
    c_tilde = tanh(zc)

    print("\nCandidate Cell:")
    print("zc = (0.7*",xt,") + (0.3*",round(h,4),") =",round(zc,4))
    print("c~ = tanh(",round(zc,4),") =",round(c_tilde,4))

    # Cell State
    c_new = f*c + i*c_tilde

    print("\nCell State Update:")
    print("c =",round(f,4),"*",round(c,4),"+",round(i,4),"*",round(c_tilde,4))
    print("c =",round(c_new,4))

    c = c_new

    # Output Gate
    zo = Wo*xt + Who*h
    o = sigmoid(zo)

    print("\nOutput Gate:")
    print("zo = (0.8*",xt,") + (0.4*",round(h,4),") =",round(zo,4))
    print("o = sigmoid(",round(zo,4),") =",round(o,4))

    # Hidden State
    h_new = o * tanh(c)

    print("\nHidden State:")
    print("h =",round(o,4),"* tanh(",round(c,4),")")
    print("h =",round(h_new,4))

    h = h_new

# prediction
Wy = 4
y_hat = Wy*h

print("\n====================")
print("Final Prediction")
print("y = 4 *",round(h,4))
print("y =",round(y_hat,3))
```
